### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data Preprocessing

```python
df = pd.read_csv('IRIS.csv')
df.head(n=10)
```

```python
X = df.values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
X[:,-1] = le.fit_transform(X[:,-1])   # Encoding the string labels into numerical labels
```

```python
print(X[:10,:])
```

```python
# Plotting the Dataset by just taking 2 features in consideration
plt.scatter(X[:,0],X[:,1],c=X[:,-1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

#### Using Elbow-Method for knowing best values of no. of clusters(k)

```python
k = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X[:,:-1])
    k.append(kmeans.inertia_)
```

```python
plt.plot(k)
plt.show()
```

#### For k = 2(Best value of k according to Elbow-Method)

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X[:,:-1])
```

```python
kmeans.labels_
```

```python
# Plotting the Dataset by just taking 2 features in consideration
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker="^",color="red")
plt.show()
```

#### For k = 3(Original Number of clusters)

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X[:,:-1])
```

```python
kmeans.labels_
```

```python
# Plotting the Dataset by just taking 2 features in consideration
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker="^",color="red")
plt.show()
```
